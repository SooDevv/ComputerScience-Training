# Binary Search Tree 
- 2가지 특징
    * 어떤 특정 노드를 선택했을 떄 그 노드 기준으로 왼쪽 서브 트리에 존재하는 노드의 모든 데이터는 기준 노드의 값보다 작고, 오른쪽 서브트리에 있는 노드의 모든 데이터는 기준 노드보다 크다
    * 이진 탐색 트리는 중복된 값을 가질 수 없다
- 추상 자료형
    * BST.insert(data) --> None
    * BST.search(target) --> node
    * BST.remove(target) --> node
    * BST.insert_node(node) --> None : 노드를 삽입 
- 이진 트리의 일종(IS-A)이므로 이진 트리 클래스를 상속

```python
class TreeNode:
    def __init__(self):
        self.__data = None
        self.__left = None
        self.__right = None
        
    def __del__(self):
        print("TreeNode of {} is deleted".format(self.data))
    
    @property
    def data(self):
        return self.__data
    
    @data.setter
    def deta(self, data):
        self.__data = data
    
    @property
    def left(self):
        return self.__left
    
    @left.setter
    def left(self, left):
        self.__left = left
    
    @property
    def right(self):
        return self.__right

    @right.setter
    def right(self, right):
        self.__right = right
```

```python
# 이진 트리와 같음 
class BST:
    def __init__(self):
        self.root = None
    
    def get_node(self):
        return self.root
    
    # 전위 순회 --> root, left, right
    # f = lambda a: print(a, end= " ")
    def preorder_traverse(self, cur, f):
        if not cur:
            return
        
        f(cur.data) # just print
        self.preorder_traverse(cur.left, f)
        self.preorder_traverse(cur.right, f)
        
    ## Insert method --> return None
    def insert(self, data):
        new_node = TreeNode()
        new_node.data = data
        
        cur = self.root
        if cur == None:
            self.root = new_node
            return
        
        # 삽입할 노드의 위치를 찾아 삽입
        while True:
            parent = cur
            
            # 왼쪽으로
            if data < cur.data:
                cur = cur.left
                if not cur:
                    parent.left = new_node
                    return
            # 오른쪽으로
            else:
                cur = cur.right
                if not cur:
                    parent.right = new_node
                    return
                
    ## Search method --> return node 
    def search(self, target):
        cur = self.root
        
        while cur:
            if target == cur.data:
                return cur
            
            elif target < cur.data:
                cur = cur.left
            else:
                cur = cur.right
                
        return cur
        
                
```
