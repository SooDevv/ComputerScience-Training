## 객체 지향으로 은행 입출금 프로그램 만들기 
은행 <br>
- 계좌 클래스를 관리하는 은행은 계좌가 총 몇 개 개설되었는지 알고싶음
- 개설된 계좌 수를 클래스 멤버로 두어 인스턴스를 만들 떄 마다 하나씩 증가시킴<br>

고객<br>
- 하지만 고객은 계좌 수를 인스턴스 멤버로 가질 필요가 없다. <br>
- 고객의 이름이나 잔액처럼 객체마다 다른 데이터만 가지면 된다. 

```python
class Account:
    # 클래스 변수
    num_acnt = 0
    
    @classmethod
    def get_num_acnt(cls):
        """
        총 계좌 수
        """
        return cls.num_acnt
    
    def __init__(self, name, money):
        self.user = name
        self.balance = money
        Account.num_acnt += 1 #계좌 생성될 때 마다(객체 생성될 때 마다) 총 계좌수 +1
    
    def deposit(self, money): # 예금
        if money < 0: # 예금한 돈이 없다면 
            return
        self.balance += money
    
    def withdraw(self, money): # 출금
        if money > 0 and self.balance >= money:
            self.balance -= money
            return money
        else:
            return None
    def transfer(self, other, money):# 계좌 이체
        """
        obj(other, money) -> bool
        other : 타인
        money : 이체 금액
        return True if the balance is enough to transfer
        return False if not
        """
        mon = self.withdraw(money) # step1. 자기 계좌에서 돈 뻄
        if mon: # 돈이 있다면
            other.deposit(mon) # step2. 타인 계좌에 입금
            return True # step3. 계좌 이체 성공
        else:
            return False
    
    def __str__(self):
        return 'User: {}, balance : {}'.format(self.user, self.balance)
        
```

```python
if __name__=="__main__":
    # 객체 생성
    my_acnt = Account('soojung', 10000)
    clova_acnt = Account('clova', 5000)
    
    # 생성 확인
    print('--------- 계좌 생성 확인 ---------')
    print(my_acnt)
    print(clova_acnt)
    print()
    
    # 인스턴스 메서드를 호출하는 방법 -> 2가지
    # N1. by object 
    my_acnt.deposit(500)
    # N2 . by class
    # Account.deposit(my_acnt, 500)
    
    # Deposit 확인
    print('--------- deposit ---------')
    print(my_acnt , "\n")
    
    # Withdraw
    print('--------- withdraw ---------')
    money = my_acnt.withdraw(1500)
    if money:
        print('withdrawn money : {}'.format(money))
    else:
        print('Not enough to draw')
    print()
    
    #클래스 멤버에 접근하는 방법 -> 2가지
    print('--------- class member ---------')
    # N1. by class
    print(Account.num_acnt)
    # N2. by object 
    # print(my_acnt.num_acnt)
    print()
    
    # 클래스 메서드를 호출하는 방법 -> 2가지
    print('--------- class method ---------')
    # N1. by class
    n_cnt = Account.get_num_acnt()
    # N2. by object
    # n_cnt = my_acnt.get_num_acnt()
    
    print('------------------------------------')
    print('총 계좌 수 : {}'.format(n_cnt))
    print('------------------------------------')
    
    #메시지 패싱
    print('---------message passing---------')
    print(my_acnt)
    print(clova_acnt)
    res = my_acnt.transfer(clova_acnt,5000)
    if res:
        print('>> 이체가 완료되었습니다.')
    else: 
        print('>> 잔액이 부족합니다.')
    print('------------------------------------')
    print(my_acnt)
    print(clova_acnt)
    
```
